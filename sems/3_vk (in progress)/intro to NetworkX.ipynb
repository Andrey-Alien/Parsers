```python
import networkx as nx           # Пакет для работы с графами
import matplotlib.pyplot as plt # Пакет для визуализации

# Опция для отображения картинок
%matplotlib inline 
# Размер картинок
plt.figure(figsize=(10,5))
```

```python
g = nx.Graph()
g.add_edge('a','b',weight=0.1)
g.add_edge('b','c',weight=1.5)
g.add_edge('a','c',weight=1.0)
g.add_edge('c','d',weight=2.2)
print(nx.shortest_path(g,'b','d'))
```

```python
G = nx.erdos_renyi_graph(15,0.8)
```

```python
p = nx.shortest_path(G,6,11)
p[1:]
```

```python
list(zip(p,p[1:]))
```

```python
pos = nx.spring_layout(G)
pos
```

```python
nx.draw_networkx_nodes(G,pos,node_color='#6E8EBD',node_size=500,linewidths=2.0);
nx.draw_networkx_labels(G,pos,font_size=18,font_color='w');
nx.draw_networkx_edges(G,pos,width=2.0,style='dotted');
nx.draw_networkx_edges(G,pos,edgelist=list(zip(p,p[1:])),width=2.0,edge_color='r');
plt.axis('off'); None
```
