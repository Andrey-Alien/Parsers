```python
# Функция, которая всё достаёт по одному запросу
def tweeter_20(polzovatel,nmax):
    mainpage = 'https://twitter.com/i/profiles/show/'+polzovatel+'/timeline/tweets?include_available_features=1&include_entities=1&max_position='+nmax+'&reset_error_state=false'
    response = requests.get(mainpage)
    dic = response.json()
    s = dic['items_html']
    soup = BeautifulSoup(s,"lxml")
    
    twits = soup.findAll("div", { "class" : "js-tweet-text-container" })
    tweets = [item.text for item in twits]
    
    data = soup.findAll("a", { "class" : "tweet-timestamp js-permalink js-nav js-tooltip" })
    date = []
    repost=[]
    dcid=[]
    for item in data:
        spl = re.split('data-conversation-id="|" href="|" title="|data-time="|" data-time-ms=',str(item))
        dcid.append(spl[1])
        repost.append(spl[2])
        date.append(datetime.datetime.fromtimestamp(int(spl[4])).strftime('%Y-%m-%d %H:%M:%S'))
    
    
    df={'Твиты': tweets,'Даты':date,'Автор':repost,'id':dcid}
    return(df)
```

```python
df = tweeter_20('RT_russian','999999999999999999')
```
