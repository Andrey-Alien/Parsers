```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup
def getTitle(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        return None
    try:
        bsObj = BeautifulSoup(html.read())
        title = bsObj.body.h1
    except AttributeError as e:
        return None
    return title
title = getTitle("http://www.pythonscraping.com/exercises/exercise1.html")
if title == None:
    print("Title coul not be found")
else:
    print(title)
```

Код выше выдаёт кучу разных ссылок, которые являются мусорными. Нужны все ссылки на статьи, которые имеют особую структуру. Находим эту структуру и записываем её с помощью регулярных выражений. Мы аеликолепны. 

```python
from urllib.request import urlopen 
from bs4 import BeautifulSoup 
import re
html = urlopen("http://en.wikipedia.org/wiki/Kevin_Bacon")
bsObj = BeautifulSoup(html)
for link in bsObj.find("div", {"id":"bodyContent"}).findAll("a",href=re.compile("^(/wiki/)((?!:).)*$")):
    if 'href' in link.attrs:
        print(link.attrs['href'])
```

Преобразуем всё это дело в две функции: первая выдаёт активные ссылки со статьи, вторая ходит по ним. 

```python
# Бродилка по ссылкам :) Рандомная бродилка по ссылкам! Хороший краулер.
# По аналогии сделать для сбора имён с твиттера! 

from urllib.request import urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re
random.seed(datetime.datetime.now())
def	getLinks(articleUrl):
    html = urlopen("http://en.wikipedia.org"+articleUrl)
    bsObj = BeautifulSoup(html)
    return bsObj.find("div", {"id":"bodyContent"}).findAll("a",href=re.compile("^(/wiki/)((?!:).)*$"))

links = getLinks("/wiki/Kevin_Bacon")
while len(links) > 0:
    newArticle = links[random.randint(0, len(links)-1)].attrs["href"]
    print(newArticle)
    links = getLinks(newArticle)
```

```python
# Краулеры, сбор информации и такое прочее. Важно не дать краулеру повторно проходить по уже обследованным ссылкам. 
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
pages = set()
def getLinks(pageUrl):
    global pages
    html = urlopen("http://en.wikipedia.org"+pageUrl)
    bsObj = BeautifulSoup(html)
    for link in bsObj.findAll("a", href=re.compile("^(/wiki/)")):
        if 'href' in link.attrs:
            if link.attrs['href'] not in pages:
                #We have encountered a new page
                newPage = link.attrs['href']
                print(newPage)
                
                pages.add(newPage)
                getLinks(newPage)
getLinks("")
```

### Twitter 

Компания защищает свой API и это довольно справедливо (наверное). С 230 миллионами активных пользователей и выручкой в 100 миллионов долларов в месяц компания не хочет допустить утечки своих данных кому попало в больших количиствах. Слишком много желающих собрать данные могут причинить вред сайту.

На API накладывается ограничение в количестве обращений, которое юзер может сделать. Ограничения делятся на две категории от 15 запросов за 15-минутный период и 180 запросов за 15-минутный период в зависимости от типа вызова. Например, можно сделать 12 вызовов в минуту для сбора базовой информации о пользователях сети, но только один запрос в минуту для сбора подписчиков.
